```python
from bs4 import BeautifulSoup
import requests
```

```python
response = requests.get("http://www.statelocalgov.net/")
print("Status code: " + str(response.status_code))
print("Page HTML:\n")
print(response.text[:1000])
```

```python
import re

html = response.text
soup = BeautifulSoup(html, "html.parser")
links = soup.find_all("a")
state_links = []
for a in links:
    is_state_link = a.has_attr('href') and re.search(r"^local-.*\.cfm", a['href'])
    if is_state_link:
        state_links.append((a['href'], a.text))
print(state_links)
```

```python
cities = []
counties = []
for state_link, state_name in state_links:
    soup = BeautifulSoup(requests.get("http://www.statelocalgov.net/" + state_link).text, "html.parser")
    
    for county_tag in soup.find_all("h3")[1:]:
        counties.append({
            "name": county_tag.text,
            "state": state_name
        })
        for city_tag in county_tag.nextSibling.find_all("li"):
            cities.append({
                "name": city_tag.a.text,
                "link": city_tag.a.get("href"),
                "county": county_tag.text,
                "state": state_name
            })
```

```python
for c in cities:
    print(c['name'], c['county'], c['state'], c['link'])
```
